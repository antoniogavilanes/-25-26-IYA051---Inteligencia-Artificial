```python
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Definir la función teórica
def f(x):
    """Función teórica para generar los puntos"""
    return x * np.sin(x)

# Generar los puntos X usados para representar la función
x_plot = np.linspace(0, 10, 100)

# Generar de forma aleatoria la muestra de puntos x y las etiquetas y
x = np.linspace(0, 10, 100)
rng = np.random.RandomState(0)
rng.shuffle(x)
x = np.sort(x[:20])
y = f(x)

# Definir colores y anchura de la línea
colors = ['red', 'orange', 'green']
lw = 2

# Dibujar la gráfica teórica f(x) y los puntos de entrenamiento
plt.plot(x_plot, f(x_plot), color='blue', linewidth=lw, label="Función teórica")
plt.scatter(x, y, color='navy', s=30, marker='o', label="Puntos de entrenamiento")
plt.title("Función teórica y puntos de entrenamiento")
plt.legend()
plt.show()

# Realizar la regresión polinómica usando la función polyfit de numpy
print("Ajuste de Regresión Polinómica")

# Dibujar nuevamente la gráfica teórica y los puntos de entrenamiento
plt.plot(x_plot, f(x_plot), color='blue', linewidth=lw, label="Función teórica")
plt.scatter(x, y, color='navy', s=30, marker='o', label="Puntos de entrenamiento")

# Realizar el ajuste polinómico para grados 3, 4 y 5 y calcular el ECM
for count, degree in enumerate([3, 4, 5]):
    # Ajustar el polinomio de grado 'degree' a los datos de entrenamiento x, y
    coeffs = np.polyfit(x, y, deg=degree)
    p = np.poly1d(coeffs)

    # Mostrar la ecuación del polinomio
    print(f"Polinomio de grado {degree}:")
    print(p)
    print("")

    # Predecir los valores de y para el conjunto de entrenamiento
    y_pred = np.polyval(p, x)

    # Calcular el error cuadrático medio (ECM)
    ecm = np.mean((y - y_pred)**2)
    print(f"Error cuadrático medio (ECM) para grado {degree}: {ecm}")
    print("")

    # Dibujar la gráfica del polinomio ajustado
    y_plot = np.polyval(p, x_plot)
    plt.plot(x_plot, y_plot, color=colors[count], linewidth=lw, label=f"Grado {degree}")

# Leyenda y mostrar gráfico
plt.legend(loc='lower left')
plt.title("Ajuste de Regresión Polinómica")
plt.show()

# Utilizar el modelo de regresión polinómica para predecir valores
# Predicción para un valor de X=6 con el modelo de regresión polinómica de grado 5
coeffs = np.polyfit(x, y, deg=5)
y_pred = np.polyval(np.poly1d(coeffs), 6)
print(f"Predicción para X=6: y = {y_pred}")
```

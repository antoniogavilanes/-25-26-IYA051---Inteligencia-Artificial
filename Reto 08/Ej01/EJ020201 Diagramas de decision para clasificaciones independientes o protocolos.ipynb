# DIAGRAMAS DE DECISION
de respuestas Excluyentes y no basados en datos estadísticos (grupos delimitados)

Un ejemplo de diagrama de protocolos para covid-19:


```python
# Función para preguntar con validación de respuestas
def preguntar(pregunta, opciones_validas):
    respuesta = ""
    while respuesta not in opciones_validas:
        print(pregunta)
        print(f"Opciones válidas: {', '.join(opciones_validas)}")
        respuesta = input("Escribe tu respuesta: ").strip().lower()
        if respuesta not in opciones_validas:
            print("Respuesta inválida. Inténtalo de nuevo.")
    return respuesta

# Árbol de decisiones para síntomas y aislamiento
def arbol_decision_covid():
    print("Bienvenido al árbol de decisiones de COVID-19 para síntomas y aislamiento.")

    # Primera pregunta: ¿Tiene síntomas?
    tiene_sintomas = preguntar("¿Tienes síntomas de COVID-19?", ["si", "no"])

    if tiene_sintomas == "si":
        # Pregunta sobre síntomas
        sintomas_graves = preguntar("¿Presentas síntomas graves como dificultad para respirar o dolor en el pecho?", ["si", "no"])
        if sintomas_graves == "si":
            print("Debes buscar atención médica de inmediato. Aislate y sigue las indicaciones de un profesional de salud.")
        else:
            print("Debes aislarte durante 10 días desde que comenzaron los síntomas y seguir las indicaciones de salud.")
    else:
        # Si no tiene síntomas, preguntar si tuvo un contacto cercano
        contacto_cercano = preguntar("¿Has tenido contacto cercano con una persona que dio positivo en COVID-19?", ["si", "no"])
        if contacto_cercano == "si":
            vacunado = preguntar("¿Estás completamente vacunado (incluyendo refuerzos si aplican)?", ["si", "no"])
            if vacunado == "si":
                print("No necesitas hacer cuarentena, pero debes realizarte una prueba el día 5.")
            else:
                print("Debes hacer cuarentena durante 5 días. Si no presentas síntomas, puedes salir de cuarentena después de 5 días.")
        else:
            print("No es necesario realizar aislamiento ni cuarentena en este momento. Mantén las medidas preventivas habituales.")

# Ejecutar el árbol de decisiones
arbol_decision_covid()
```

```python
# Función para hacer preguntas con validación
def preguntar(pregunta, opciones_validas):
    respuesta = ""
    while respuesta not in opciones_validas:
        print(pregunta)
        print(f"Opciones válidas: {', '.join(opciones_validas)}")
        respuesta = input("Escribe tu respuesta: ").strip().lower()
        if respuesta not in opciones_validas:
            print("Respuesta inválida. Por favor, intenta de nuevo.")
    return respuesta

# Árbol de decisiones basado en la página 1
def arbol_decision_covid():
    print("Bienvenido al Árbol de Decisión para COVID-19: Trabajadores y Asistentes")

    # Pregunta 1: ¿La persona tiene síntomas?
    tiene_sintomas = preguntar("¿Tienes síntomas de COVID-19?", ["si", "no"])

    if tiene_sintomas == "si":
        # Rama para personas con síntomas
        resultado_prueba = preguntar("¿Te has hecho una prueba de COVID-19?", ["si", "no"])

        if resultado_prueba == "si":
            resultado_positivo = preguntar("¿El resultado de la prueba fue positivo?", ["si", "no"])

            if resultado_positivo == "si":
                print("Debes aislarte por al menos 10 días desde el inicio de los síntomas.")
                print("Consulta a un proveedor de salud y sigue sus indicaciones.")
            else:
                print("Si el resultado es negativo, continúa aislado hasta que los síntomas mejoren.")
        else:
            print("Debes realizarte una prueba de COVID-19 y aislarte hasta obtener el resultado.")
            print("Consulta a un proveedor de salud.")

    else:
        # Pregunta 2: ¿Has estado en contacto cercano con alguien que dio positivo en COVID-19?
        contacto_cercano = preguntar("¿Has tenido contacto cercano con una persona positiva de COVID-19?", ["si", "no"])

        if contacto_cercano == "si":
            vacunado = preguntar("¿Estás completamente vacunado con una dosis de refuerzo (si es aplicable)?", ["si", "no"])

            if vacunado == "si":
                print("No necesitas hacer cuarentena, pero deberías hacerte una prueba el día 5 después del contacto.")
            else:
                print("Debes hacer cuarentena por al menos 5 días. Hazte una prueba el día 5.")
                cuarentena_finalizada = preguntar("¿Te has hecho una prueba el día 5 y el resultado fue negativo?", ["si", "no"])
                if cuarentena_finalizada == "si":
                    print("Puedes salir de la cuarentena si no presentas síntomas.")
                else:
                    print("Si no te has hecho una prueba, debes continuar en cuarentena por 10 días.")
        else:
            print("No es necesario realizar cuarentena ni aislamiento, pero mantén las precauciones habituales.")

# Ejecutar el árbol de decisión
arbol_decision_covid()
```


Clasificador de animales según características:

Vertebrados.png



```python
# Función para preguntar y validar respuestas
def preguntar(pregunta, opciones_validas):
    respuesta = ""
    while respuesta not in opciones_validas:
        print(pregunta)
        print(f"Opciones válidas: {', '.join(opciones_validas)}")
        respuesta = input("Escribe tu respuesta: ").strip().lower()
        if respuesta not in opciones_validas:
            print("Respuesta inválida. Por favor, intenta de nuevo.")
    return respuesta

# Función principal del árbol de decisión
def clasificar_animal():
    print("Bienvenido al clasificador de animales vertebrados.")

    # Pregunta sobre el tipo de cuerpo
    cuerpo = preguntar("¿El cuerpo del animal tiene pelo, escamas, plumas o piel húmeda?",
                       ["pelo", "escamas", "plumas", "piel húmeda"])

    if cuerpo == "pelo":
        # Pregunta adicional para diferenciar mamíferos acuáticos
        acuatico = preguntar("¿El animal es acuático (como ballenas o delfines)?", ["si", "no"])
        if acuatico == "si":
            print("El animal es un mamífero acuático.")
        else:
            print("El animal es un mamífero.")

    elif cuerpo == "escamas":
        # Pregunta sobre el desplazamiento
        desplazamiento = preguntar("¿El animal nada, repta o camina?", ["nada", "repta", "camina"])
        if desplazamiento == "nada":
            print("El animal es un pez.")
        else:
            print("El animal es un reptil.")

    elif cuerpo == "plumas":
        print("El animal es un ave.")

    elif cuerpo == "piel húmeda":
        print("El animal es un anfibio.")

    # Finalización del árbol
    print("Clasificación completada.")

# Ejecutar el árbol de decisión
clasificar_animal()
```

Python no tiene una estructura de control switch o case como en otros lenguajes (por ejemplo, switch en C o Java). Sin embargo, a partir de Python 3.10, se introdujo una nueva estructura de control llamada match-case, que se comporta de manera similar a switch-case en otros lenguajes. Por ejemplo:

```python
def clasificar_animal(cuerpo):
    match cuerpo:
        case "pelo":
            return "El animal es un mamífero."
        case "escamas":
            return "El animal es un pez o un reptil."
        case "plumas":
            return "El animal es un ave."
        case "piel húmeda":
            return "El animal es un anfibio."
        case _:
            return "Característica desconocida."

# Ejemplo de uso
animal = input("Introduce el tipo de cuerpo del animal (pelo, escamas, plumas, piel húmeda): ").strip().lower()
print(clasificar_animal(animal))
```

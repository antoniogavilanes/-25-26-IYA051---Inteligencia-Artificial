```python
def clasificar_animal():
    print("=== Clasificación de Animales ===")

    columna = input("¿El animal tiene columna vertebral? (si/no): ").strip().lower()

    # -------------------------------------------------------------------------
    # NIVEL 1 → Vertebrado o Invertebrado
    # -------------------------------------------------------------------------
    if columna == "si":
        # VERTEBRADOS
        piel = input("Tipo de cuerpo (pelos, escamas, plumas, piel humeda): ").strip().lower()

        # ---------------------------------------------------------------------
        # NIVEL 2 → Tipo de piel
        # ---------------------------------------------------------------------
        if piel == "pelos":
            respira = input("¿Respira con pulmones? (si/no): ").strip().lower()

            # -----------------------------------------------------------------
            # NIVEL 3
            # -----------------------------------------------------------------
            if respira == "si":
                patas = input("¿Tiene patas? (si/no): ").strip().lower()

                # -------------------------------------------------------------
                # NIVEL 4
                # -------------------------------------------------------------
                if patas == "si":
                    habitat = input("Hábitat principal (tierra/agua): ").strip().lower()

                    # ---------------------------------------------------------
                    # NIVEL 5
                    # ---------------------------------------------------------
                    if habitat == "tierra":
                        return "→ Mamífero"
                    else:
                        return "→ Mamífero acuático"
                else:
                    return "→ Mamífero especial (como ballena o delfín)"
            else:
                return "→ No coincide con mamíferos"
            
        elif piel == "escamas":
            respira = input("¿Respira por branquias? (si/no): ").strip().lower()

            if respira == "si":
                return "→ Pez"
            else:
                patas = input("¿Tiene patas? (si/no): ").strip().lower()
                if patas == "si":
                    return "→ Reptil"
                else:
                    return "→ Reptil (sin extremidades, como serpiente)"

        elif piel == "plumas":
            vuela = input("¿Puede volar? (si/no): ").strip().lower()
            if vuela in ["si", "no"]:
                return "→ Ave"

        elif piel == "piel humeda":
            medio = input("¿Vive en agua, tierra o ambos? ").strip().lower()
            if medio in ["agua", "tierra", "ambos"]:
                return "→ Anfibio"

        return "No se pudo clasificar como vertebrado"

    # -------------------------------------------------------------------------
    # INVERTEBRADOS
    # -------------------------------------------------------------------------
    else:
        proteccion = input("¿Tiene protección corporal? (si/no): ").strip().lower()

        # ---------------------------------------------------------------------
        # NIVEL 2
        # ---------------------------------------------------------------------
        if proteccion == "si":
            tipo = input("Tipo de protección (cubierta dura / concha / espinas): ").strip().lower()

            # -----------------------------------------------------------------
            # NIVEL 3
            # -----------------------------------------------------------------
            if "cubierta" in tipo:
                patas = input("¿Cuántas patas tiene? ").strip().lower()

                # -------------------------------------------------------------
                # NIVEL 4
                # -------------------------------------------------------------
                if "6" in patas:
                    return "→ Artrópodo: Insecto"
                elif "8" in patas:
                    return "→ Artrópodo: Arácnido"
                elif "muchas" in patas:
                    return "→ Artrópodo: Miriápodo"
                else:
                    return "→ Artrópodo: Crustáceo"

            elif "concha" in tipo:
                externa = input("¿Tiene concha externa? (si/no): ").strip().lower()

                if externa == "si":
                    piezas = input("¿La concha es de 1 o 2 piezas? ").strip().lower()

                    if "1" in piezas:
                        return "→ Molusco: Gasterópodo"
                    elif "2" in piezas:
                        return "→ Molusco: Bivalvo"
                else:
                    return "→ Molusco: Cefalópodo"

            elif "espinas" in tipo:
                forma = input("¿Simetría radial? (si/no): ").strip().lower()
                if forma == "si":
                    return "→ Equinodermo"

        # ---------------------------------------------------------------------
        # SIN PROTECCIÓN CORPORAL
        # ---------------------------------------------------------------------
        else:
            cuerpo = input("¿El cuerpo es segmentado? (si/no): ").strip().lower()

            # -----------------------------------------------------------------
            # NIVEL 3
            # -----------------------------------------------------------------
            if cuerpo == "si":
                anillos = input("¿Tiene anillos visibles? (si/no): ").strip().lower()

                if anillos == "si":
                    return "→ Gusano: Anélido"
                else:
                    return "→ Gusano: Nemátodo"

            else:
                forma = input("¿Cuerpo con forma de plancha o transparente? ").strip().lower()

                if "plana" in forma:
                    return "→ Platelminto"
                elif "transparente" in forma:
                    tentaculos = input("¿Tiene tentáculos? (si/no): ").strip().lower()

                    if tentaculos == "si":
                        tipo = input("¿Flota o está fija? (flota/fija): ").strip().lower()
                        if tipo == "flota":
                            return "→ Celentéreo: Medusa"
                        else:
                            return "→ Celentéreo: Pólipo"
                else:
                    return "→ Porífero (esponja)"

        return "No se pudo clasificar como invertebrado"


# EJECUCIÓN
resultado = clasificar_animal()
print("\nClasificación final:", resultado)
```

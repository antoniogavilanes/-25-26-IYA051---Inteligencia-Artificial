```python
# ---------------------------------------------
# CLASIFICACIÃN DE VINOS CON ÃRBOL DE DECISIÃN
# ---------------------------------------------

# 1. Importar librerÃ­as
import sys
!{sys.executable} -m pip install scikit-learn
!{sys.executable} -m pip install matplotlib

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# 2. Cargar el dataset Wine de scikit-learn
data = load_wine()

X = data.data                 # CaracterÃ­sticas (13 variables quÃ­micas)
y = data.target               # Etiqueta (Clase 0, 1 o 2)

print("CaracterÃ­sticas del dataset:", data.feature_names)
print("Clases:", data.target_names)
print("\nPrimeras 5 muestras:\n", X[:5])

# 3. DivisiÃ³n del dataset: 70% entrenamiento / 30% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

# 4. Entrenamiento del modelo con Ãrbol de DecisiÃ³n
clf = DecisionTreeClassifier(random_state=1)
clf.fit(X_train, y_train)

# 5. VisualizaciÃ³n del Ãrbol de DecisiÃ³n
plt.figure(figsize=(20, 12))
tree.plot_tree(
    clf,
    filled=True,
    feature_names=data.feature_names,
    class_names=data.target_names
)
plt.title("Ãrbol de DecisiÃ³n â ClasificaciÃ³n de Vinos")
plt.show()

# 6. EvaluaciÃ³n del modelo
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("\nPrecisiÃ³n del modelo:", accuracy)
```
